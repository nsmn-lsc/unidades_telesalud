```python
import pandas as pd

file = "files/Especificaciones de las Unidades de Telesalud y Telemedicina_HGO_Implementadas_281125.xlsx"
df = pd.read_excel(file)
```

# Seleccionar las columnas a utilizar

```python

df_unidades = df[["Municipio","CLUES", "Nombre de la Unidad"]]
df_unidades["Municipio"] = df_unidades["Municipio"].str.strip()

# Contar unidades por municipio (como DataFrame)
unidades_por_municipio = df_unidades.groupby('Municipio').size().reset_index(name='Cantidad de Unidades')
unidades_por_municipio = unidades_por_municipio.sort_values('Cantidad de Unidades', ascending=False)

conteo = df_unidades.groupby("Municipio")["CLUES"].count()
print(conteo)

```

```python
import plotly.express as px

# Gráfico de barras horizontal
fig = px.bar(unidades_por_municipio, 
             y='Municipio', 
             x='Cantidad de Unidades',
             title='Unidades de Telesalud por Municipio en Hidalgo',
             orientation='h',
             color='Cantidad de Unidades',
             color_continuous_scale='Cividis',
             text='Cantidad de Unidades')

fig.update_traces(textposition='outside')
fig.update_layout(
    height=1200,  # Aumentar altura para más municipios
    yaxis={
        'categoryorder':'total ascending',
        'tickfont': {'size': 10}  # Reducir tamaño de fuente
    },
    xaxis_title='Cantidad de Unidades',
    yaxis_title='Municipio',
    showlegend=False,
    margin=dict(l=150, r=50, t=80, b=50)  # Más margen izquierdo para nombres largos
)

fig.show()
```
